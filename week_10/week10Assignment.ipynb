Apply tSNE to the digits dataset (transform to 2D) and perform K-means clustering into 10 clusters in 2D space. calculate sklearn.metrics.adjusted_rand_score between the labels obtained in clustering and the original labels. http://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_digits.html

```python
digits = load_digits()
X = digits.data
y = digits.target
```

```python
def scatterplot_2D(data, title, labels=None):
    """ 
    Plotting function taken from Alex's lecture
    """
    assert(data.shape[1] == 2)
    # class labels are turned into colors
    if labels is None:
        c = 'black' # default of points is black
    else:
        color_scale = np.linspace(0, 1, len(set(labels)))
        c = [plt.cm.Set2(color_scale[i]) for i in labels]

    fig = plt.figure()
    ax = fig.add_subplot(111)
    # ax.patch.set_facecolor('grey')
    ax.scatter(data[:, 0], data[:, 1], color = c, alpha=0.75)
    ax.axis('square')
    ax.set_xlabel('R1')
    ax.set_ylabel('R2')
    plt.suptitle(title)
    plt.show()
```

```python
digits_TSNE = TSNE(n_components=2).fit_transform(X)
# n_components: dimension of the embedded space.
```

Other manifold learning methods:

```python
digits_MDS = MDS(n_components=2).fit_transform(X)
```

```python
digitsISO = Isomap(n_components=2).fit_transform(X) 
# http://scikit-learn.org/stable/modules/generated/sklearn.manifold.Isomap.html
```

```python
scatterplot_2D(digits_TSNE, "Digits TSNE", y)
scatterplot_2D(digitsISO, "Digits Isomap", y)
scatterplot_2D(digits_MDS, "Digits MDS", y)
```

And just for fun:

```python
digits_PCA = PCA(n_components = 5).fit_transform(X)
digits_TSNE_test = TSNE(n_components = 2).fit_transform(digits_PCA)
digits_MDS_test = TSNE(n_components = 2).fit_transform(digits_PCA)
```

```python
scatterplot_2D(digits_TSNE_test, "Digits PCA > TSNE", y)
scatterplot_2D(digits_MDS_test, "Digits PCA > MDS", y)
# Get better separation on MDS, but not sure if it means anything
```

Now for Kmeans clustering

```python
model = KMeans(n_clusters = 10, random_state = 1)
predicted_clusters = model.fit_predict(X)
predicted_cluster_reduced = model.fit_predict(digits_TSNE)
score = adjusted_rand_score(y, predicted_clusters)
score_reduced = adjusted_rand_score(y, predicted_cluster_reduced)
print("The ARI of Kmeans on the original data is {0}, and the ARI for Kmeans on data reduced by TSNE is {1}".format(score, score_reduced))
```

Use BRCA gene (protein) expression data provided in this class. Apply K-means clustering to the original expression data and to the result of NMF decomposition with 10 components (H matrix). For simplicity, use k=10 as the number of clusters in K-means. Calculate adjusted_rand_score as a measure of consensus between the two resulting sets of clustering labels.

NMF decomposition. 

From http://scikit-learn.org/stable/modules/decomposition.html#nmf

NMF [1] is an alternative approach to decomposition that assumes that the data and the components are non-negative. NMF can be plugged in instead of PCA or its variants, in the cases where the data matrix does not contain negative values. It finds a decomposition of samples X into two matrices W and H of non-negative elements, by optimizing the distance d between X and the matrix product WH. 

```python
expression_data = 'https://dcc.icgc.org/api/v1/download?fn=/release_18/Projects/BRCA-US/protein_expression.BRCA-US.tsv.gz'
data = pd.read_csv(expression_data, delimiter='\t')
X = data
(X.head)
```

```python


# Process data: 
data = data[['icgc_donor_id', 'gene_name', 'antibody_id', 'normalized_expression_level']]
data_mean = data.groupby(['icgc_donor_id','gene_name']).mean() 
# Unfortunately don't know why mean is taken, but alex did it :(
data_mean_wide = data_mean.unstack()

# Run Kmeans:
model = KMeans(n_clusters = 10, random_state = 1)
predicted_clusters = model.fit_predict(data_mean_wide)

# Reduce data and run Kmeans:
X_reduced = sklearn.preprocessing.MinMaxScaler().fit_transform(data_mean_wide)
nmf = NMF(n_components = 10, random_state = 1).fit(X_reduced)
W = nmf.components_
H = nmf.transform(X_reduced)
model = KMeans(n_clusters = 10, random_state = 1)
predicted_clusters_reduced = model.fit_predict(H)

# Calculate ARI between Kmeans on reduced and non reduced data
score = adjusted_rand_score(predicted_clusters_reduced, predicted_clusters)
print("ARI:", score)
# Don't really know what this tells me. That the clusters are not very similar? 
# From http://scikit-learn.org/stable/modules/generated/sklearn.metrics.adjusted_rand_score.html:
#"The Rand Index computes a similarity measure between two clusterings by considering all pairs of samples 
# and counting pairs that are assigned in the same or different clusters in the predicted and true clusterings."
# So it's a comparison of clusters
```
